```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv('/content/drive/My Drive/전산통계/brain_size.csv', sep=';', na_values='.')
```

```
data.head()
```

```
data.fillna(data.mean())
```

```
from statsmodels.formula.api import ols

model = ols('VIQ ~ Gender + Weight + Height + MRI_Count', data).fit()
print(model.summary())
```

Gender 변수의 coef가 10.15로 남자일 경우 VIQ가 10정도 높다는 것을 의미하는데 Gender 변수에 대한 p-value가 0.358로 높게 나왔다. 즉, 성별에 따라 VIQ가 영향을 받지 않는 것으로 볼 수 있다. 
MRI_Count의 p-value값은 유의미 하지만 coef의 값이 작아 큰 의미가 없어 보인다. 

```
model2 = ols('VIQ ~ FSIQ + PIQ', data).fit()
print(model2.summary())
```

FSIQ가 높은 사람이 VIQ가 1.69정도 높고 PIQ가 높은 사람이 VIQ가 0.87정도 낮은 것으로 볼 수 있다. 

VIQ^(hat) = 17.6592 + 1.6915 * FSIQ - 0.8755 * PIQ
